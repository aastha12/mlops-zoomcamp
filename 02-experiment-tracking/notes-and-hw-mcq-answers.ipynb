### Breakdown of "mlruns" folder:

- If you don't set any backend store (like sqlLite db) then by default your local system is the backend store
- There will be a folder called "0" in the mlruns folder which is the default experiment folder.
- Each experiment can have multiple runs. If you have a python file called abc.py which has mlflow autologging in it, everytime you run the python file a new run (with a new run id) is created in that experiement.
- Once you create a new experiment, you will see a new folder in the mlruns folder. This new folder's name will be the experiment id.
- The new experiment folder will contain:

    - Artifacts folder: Artifacts are model weights or preprocessors that you want to save
    - params folder: parameters like training and validation data path are saved here
    - metrics folder: contains metrics of the model like rmse
    - tags folder: if you have defined any tags like name of the developer working on the project then that gets saved here. Tags are just useful information that helps you understand your experiment better.
